# Dictionaries
- Dictionary is an unordered collection of key value pairs.
- In python, dictionaries are defined within braces{} with ech item being a pair in the form key:value. Key and value can be of any type.
- To declare a dictionary, you write dictionaryName= {dictionary key:data}


```python
my_dict = {'key1':'value1','key2':'value2', 'key3':'value3','key4':'value4'}
my_dict
```

```python
type(my_dict)
```

```python
len(my_dict)
```
